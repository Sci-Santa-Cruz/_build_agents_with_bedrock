### Introducción a los Agentes
Los Agentes son sistemas inteligentes o componentes que utilizan Modelos de Lenguaje Grandes (LLMs) para realizar tareas de manera dinámica y autónoma. Aquí está un desglose de los conceptos clave:

#### ¿Qué son los Agentes?
- **Pensamiento paso a paso**: Los agentes aprovechan los LLMs para pensar y razonar sobre problemas de manera estructurada, a menudo denominada razonamiento en cadena de pensamientos. Esto les permite planificar, evaluar y ejecutar tareas de manera efectiva.
- **Acceso a Herramientas**: Los agentes pueden utilizar herramientas externas (por ejemplo, calculadoras, bases de datos, APIs) para mejorar su capacidad de toma de decisiones y resolución de problemas.
- **Acceso a Memoria**: Los agentes pueden almacenar y recuperar contexto, lo que les permite trabajar en tareas a lo largo del tiempo, adaptarse a las interacciones con el usuario y manejar flujos de trabajo complejos.

### Características clave de los Agentes de IA
- **Percepción**: La capacidad de recopilar información del entorno a través de sensores o entradas de datos.
- **Toma de decisiones**: El uso de algoritmos de IA para procesar la información y determinar la mejor acción a tomar.
- **Acción**: La capacidad de ejecutar decisiones e interactuar con el entorno o los usuarios.
- **Aprendizaje**: La capacidad de mejorar el rendimiento a lo largo del tiempo mediante la experiencia y la retroalimentación.
- **Autonomía**: Operar de manera independiente hasta cierto punto, sin intervención humana constante.
- **Orientación a objetivos**: Trabajar hacia objetivos o tareas específicas.

El flujo entre estos pasos es gestionado por LangGraph, que maneja las transiciones de estado y asegura que cada paso se ejecute en el orden correcto.


## Conceptos Básicos de LangGraph

### Componentes Clave

#### StateGraph
- Este objeto encapsula el grafo que se está recorriendo durante la ejecución.
- Es el núcleo de nuestra aplicación, definiendo el flujo de nuestro Planificador de Viajes.
- **PlannerState**: Un tipo personalizado que representa el estado de nuestro proceso de planificación.

#### Nodos

En LangGraph, los nodos son típicamente funciones de Python. 
Hay dos nodos principales que usaremos en nuestro grafo:

1. **Nodo del agente**: Es responsable de decidir qué acciones tomar (si es que se deben tomar).
2. **Nodo de herramienta**: Este nodo orquesta la llamada a la herramienta respectiva y devuelve la salida.

#### Aristas

Las aristas definen cómo se enruta la lógica y cómo el grafo decide cuándo detenerse.
Estas definen cómo trabajan los agentes y cómo los diferentes nodos se comunican entre sí.
Algunos tipos clave de aristas son:

- **Aristas normales**: Van directamente de un nodo a otro.
- **Aristas condicionales**: Llaman a una función para determinar a qué nodo(s) ir a continuación.
- **Punto de entrada**: Define qué nodo se debe llamar primero cuando llega la entrada del usuario.
- **Punto de entrada condicional**: Llama a una función para determinar qué nodo(s) se debe llamar primero cuando llega la entrada del usuario.

#### Integración con LLM
- Utilizando un modelo de lenguaje para generar el itinerario final.

#### Integración de Memoria

````python
import os
from typing import TypedDict, Annotated, List
from langgraph.graph import StateGraph, END
from langchain_core.messages import HumanMessage, AIMessage
from langchain_core.prompts import ChatPromptTemplate
from langchain_core.runnables.graph import MermaidDrawMethod
from IPython.display import display, Image
from dotenv import load_dotenv
import os
#load_dotenv()
````

````python
class PlannerState(TypedDict):
    messages: Annotated[List[HumanMessage | AIMessage], "The messages in the conversation"]
    itinerary: str
    city: str
    user_message: str
````


## Configuración del Modelo de Lenguaje y los Prompts



````python
from langchain_aws import ChatBedrockConverse
import boto3
from langgraph.checkpoint.memory import MemorySaver
from langchain_core.runnables.config import RunnableConfig
from langchain_core.chat_history import InMemoryChatMessageHistory
from langchain_core.output_parsers import StrOutputParser
from langchain_core.prompts import ChatPromptTemplate
from langchain_core.runnables.history import RunnableWithMessageHistory
from langchain_core.prompts import ChatPromptTemplate, MessagesPlaceholder

# ---- ⚠️ Actualiza la región para tu configuración de AWS ⚠️ ----
region = 'us-east-2'

bedrock = boto3.client(
    service_name='bedrock-runtime',
    region_name=region,
)

bedrock_client = boto3.client("bedrock-runtime", region_name=region)

model_id = "us.anthropic.claude-3-5-haiku-20241022-v1:0"

llm = ChatBedrockConverse(
    model=model_id,
    temperature=0,
    max_tokens=5000,
    client=bedrock_client,
)

itinerary_prompt = ChatPromptTemplate.from_messages([
    ("system", """Eres un asistente de viaje útil. Crea un itinerario para un viaje de un día en {city} basado en los intereses del usuario.
    Sigue estas instrucciones:
    1. Usa la conversación de chat a continuación y la última entrada del humano para obtener los intereses del usuario.
    2. Siempre ten en cuenta el tiempo de viaje y los tiempos de comida: si no es posible hacer todo, dilo.
    3. Si el usuario no ha mencionado una época del año o temporada, asume que es verano en {city} y menciona esta suposición en tu respuesta.
    4. Si el usuario no ha mencionado un presupuesto de viaje, asume una cantidad razonable en dólares y menciona esta suposición en tu respuesta.
    5. Proporciona un itinerario breve y en viñetas en orden cronológico con las horas específicas del día."""), 
    MessagesPlaceholder("chat_history"),
    ("human", "{user_message}"),
])
````



## **Definir los nodos y las aristas**  


Estamos añadiendo los nodos, las aristas, así como una memoria persistente al **StateGraph** antes de compilarlo.

1. **Planes de viaje del usuario**  
2. **Invocar con Bedrock**  
3. **Generar el plan de viaje para el día**  
4. **Capacidad de agregar o modificar el plan**



````python
def input_interests(state: PlannerState) -> PlannerState:
    user_message = state['user_message']  # input("Tu mensaje: ")
    # print(f"Vamos a :: {user_message}:: para el viaje a {state['city']} basado en tus intereses mencionados en el mensaje....")

    if not state.get('messages', None):
        state['messages'] = []
    return {
        **state,
    }

def create_itinerary(state: PlannerState) -> PlannerState:
    response = llm.invoke(itinerary_prompt.format_messages(city=state['city'],
                         user_message=state['user_message'],
                         chat_history=state['messages']))
    print("\nItinerario Final:")
    print(response.content)
    return {
        **state,
        "messages": state['messages'] + [HumanMessage(content=state['user_message']), AIMessage(content=response.content)],
        "itinerary": response.content
    }
````


## **Crear y compilar el gráfico**

Ahora crearemos nuestro flujo de trabajo en LangGraph y lo compilaremos.

Primero, inicializamos un **StateGraph** con la clase **State** que definimos anteriormente.
Luego, agregamos nuestros **nodos** y **aristas** (edges).
Usamos el **Nodo INICIO** (*START Node*), un nodo especial que envía la entrada del usuario al gráfico, para indicar dónde comenzar nuestro gráfico.
El **Nodo FINAL** (*END Node*) es un nodo especial que representa un nodo terminal.


````python
# Inicializa el grafo de estados con un tipo de estado específico, en este caso, PlannerState
workflow = StateGraph(PlannerState)

# Agrega nodos al grafo de flujo de trabajo que representan diferentes etapas del proceso
# Cada nodo será una función o tarea que el grafo de estados ejecutará.
workflow.add_node("input_interests", input_interests)  # Nodo que maneja los intereses del usuario
workflow.add_node("create_itinerary", create_itinerary)  # Nodo que crea el itinerario con la información

# Establece el punto de entrada del grafo como el nodo 'input_interests'
# Esto significa que el flujo de trabajo comenzará en este nodo.
workflow.set_entry_point("input_interests")

# Agrega conexiones (aristas) entre nodos, lo que define el flujo de ejecución
# En este caso, el flujo de trabajo va de 'input_interests' a 'create_itinerary' y luego termina en el nodo END.
workflow.add_edge("input_interests", "create_itinerary")  # Después de manejar los intereses, pasa a crear el itinerario
workflow.add_edge("create_itinerary", END)  # Finalmente, el flujo termina

# El checkpointer permite que el grafo guarde su estado, persistiendo la memoria del flujo de trabajo.
# Esto es útil para poder recuperar y continuar desde el mismo punto en caso de reinicio o fallo.
memory = MemorySaver()  # Crea una instancia del MemorySaver para almacenar el estado del flujo de trabajo

# Compila el grafo de trabajo en una aplicación, pasando el checkpointer (MemorySaver) para que se guarde el estado del flujo de trabajo
app = workflow.compile(checkpointer=memory)
````

````python
# Importación de las clases necesarias para el funcionamiento de la memoria personalizada
from langgraph.store.base import BaseStore, Item, Op, Result
from langgraph.store.memory import InMemoryStore
from typing import Any, Iterable, Literal, NamedTuple, Optional, Union, cast

# Definición de la clase CustomMemoryStore que hereda de BaseStore
class CustomMemoryStore(BaseStore):
    """
    Esta clase extiende la funcionalidad de almacenamiento de la clase BaseStore
    permitiendo el uso de un almacén externo para gestionar datos.
    
    Acepta un 'ext_store' (almacén externo) que puede ser cualquier clase que
    implemente la interfaz del BaseStore.
    """

    def __init__(self, ext_store: BaseStore):
        """
        Inicializa el objeto CustomMemoryStore con un almacén externo.

        Args:
            ext_store (BaseStore): Almacén externo que se usará para la gestión de datos.
        """
        # El almacén proporcionado se guarda como un atributo de la instancia
        self.store = ext_store

    def get(self, namespace: tuple[str, ...], key: str) -> Optional[Item]:
        """
        Recupera un elemento del almacén usando un espacio de nombres y una clave.

        Args:
            namespace (tuple[str, ...]): Tupla que define el espacio de nombres para el almacenamiento.
            key (str): Clave del item que se desea obtener.

        Returns:
            Optional[Item]: El ítem recuperado del almacén, o None si no existe.
        """
        # Llama al método 'get' del almacén externo para obtener el ítem
        return self.store.get(namespace, key)

    def put(self, namespace: tuple[str, ...], key: str, value: dict[str, Any]) -> None:
        """
        Almacena un ítem en el almacén con un espacio de nombres y clave dados.

        Args:
            namespace (tuple[str, ...]): Tupla que define el espacio de nombres para el almacenamiento.
            key (str): Clave bajo la cual se almacenará el ítem.
            value (dict[str, Any]): El valor del ítem a almacenar.

        Returns:
            None: El método no devuelve nada.
        """
        # Llama al método 'put' del almacén externo para almacenar el valor
        return self.store.put(namespace, key, value)

    async def aget(self, namespace: tuple[str, ...], key: str) -> Optional[Item]:
        """
        Recupera de forma asíncrona un elemento del almacén usando un espacio de nombres y una clave.

        Args:
            namespace (tuple[str, ...]): Tupla que define el espacio de nombres para el almacenamiento.
            key (str): Clave del item que se desea obtener.

        Returns:
            Optional[Item]: El ítem recuperado del almacén, o None si no existe.
        """
        # Llama al método 'get' de forma asíncrona
        res = await self.get(namespace, key)
        return res

    async def aput(self, namespace: tuple[str, ...], key: str, value: dict[str, Any]) -> None:
        """
        Almacena de forma asíncrona un ítem en el almacén con un espacio de nombres y clave dados.

        Args:
            namespace (tuple[str, ...]): Tupla que define el espacio de nombres para el almacenamiento.
            key (str): Clave bajo la cual se almacenará el ítem.
            value (dict[str, Any]): El valor del ítem a almacenar.

        Returns:
            None: El método no devuelve nada.
        """
        # Llama al método 'put' de forma asíncrona
        res = await self.put(namespace, key, value)
        return res

    def batch(self, ops: Iterable[Op]) -> list[Result]:
        """
        Realiza múltiples operaciones de almacenamiento en batch (lote).

        Args:
            ops (Iterable[Op]): Conjunto de operaciones de almacenamiento que se desean realizar.

        Returns:
            list[Result]: Lista de resultados de cada operación realizada.
        """
        # Llama al método 'batch' del almacén externo para ejecutar las operaciones en batch
        return self.store.batch(ops)

    async def abatch(self, ops: Iterable[Op]) -> list[Result]:
        """
        Realiza múltiples operaciones de almacenamiento en batch de forma asíncrona.

        Args:
            ops (Iterable[Op]): Conjunto de operaciones de almacenamiento que se desean realizar.

        Returns:
            list[Result]: Lista de resultados de cada operación realizada.
        """
        # Llama al método 'abatch' del almacén externo para ejecutar las operaciones en batch de forma asíncrona
        res = await self.store.abatch(ops)
        return res
````



### Explicación de la clase `CustomMemoryStore`:

1. **Clase `CustomMemoryStore`**:
   - Esta clase extiende `BaseStore` y utiliza un almacén externo (`ext_store`) que puede ser un almacén en memoria o cualquier otro tipo de almacén que implemente la interfaz `BaseStore`.
   
2. **Métodos principales**:
   - `get`: Recupera un ítem del almacén externo usando un espacio de nombres y una clave.
   - `put`: Almacena un ítem en el almacén externo usando un espacio de nombres y una clave.
   - `aget` y `aput`: Son las versiones asíncronas de `get` y `put`, utilizadas cuando se manejan operaciones asíncronas.
   - `batch`: Realiza múltiples operaciones en lote de manera sincrónica.
   - `abatch`: Realiza múltiples operaciones en lote de manera asíncrona.

### Uso de las operaciones de batch:
- **Batch** permite ejecutar varias operaciones de almacenamiento (como insertar, actualizar o eliminar elementos) de manera eficiente en un solo proceso. Esto es útil para optimizar el rendimiento cuando se necesitan realizar múltiples acciones de forma consecutiva.

### Asincronía:
- Los métodos `aget` y `aput`, junto con `abatch`, permiten realizar operaciones de almacenamiento de manera asíncrona, lo que puede ser útil en aplicaciones que requieren no bloquear el hilo principal mientras se espera la operación de almacenamiento.

Este código está diseñado para ofrecer flexibilidad al trabajar con almacenamiento de memoria de manera tanto sincrónica como asincrónica.

````python
# 1. Creamos un almacén de memoria temporal.
# Usamos CustomMemoryStore que utiliza InMemoryStore internamente (es decir, guarda en RAM).
in_memory_store = CustomMemoryStore(InMemoryStore())

# 2. Definimos un espacio de nombres (namespace) para agrupar datos relacionados.
# Aquí agrupamos los mensajes de chat de un usuario.
namespace_u = ("chat_messages", "user_id_1")

# 3. Definimos una clave (key) única que identifica a un usuario específico
key_u = "user_id_1" 

# 4. Definimos el contenido que queremos guardar.
# Es un diccionario que contiene una lista con elementos 
data_to_store = {
    "data": ["hola que tal", "dame un plan a los maiamis", "agrega visita a Little Havana"]
}

# 5. Guardamos los datos en el almacén de memoria.
# Se guarda usando el namespace y la key definidos anteriormente.
in_memory_store.put(namespace_u, key_u, data_to_store)

# 6. Recuperamos el dato que acabamos de guardar, utilizando el mismo namespace y key.
item_u = in_memory_store.get(namespace_u, key_u)


````

— Ten en cuenta que no tendremos mensajes en el estado del gráfico ya que estos han sido externalizados.



````python
# Definir el estado del planificador como un TypedDict
class PlannerState(TypedDict):
    itinerary: str  # Itinerario del viaje
    city: str       # Ciudad para la que se está creando el itinerario
    user_message: str  # Mensaje original del usuario
````

````python
def input_interests(state: PlannerState, config: RunnableConfig, *, store: BaseStore) -> PlannerState:
    """
    Procesa el primer mensaje del usuario que contiene sus intereses de viaje.

    Args:
        state (PlannerState): Estado actual del flujo que contiene, entre otros, el 'user_message'.
        config (RunnableConfig): Configuración adicional de ejecución, contiene identificadores como 'thread_id'.
        store (BaseStore): Almacén de datos para persistir o consultar información (no utilizado en esta función).

    Returns:
        PlannerState: Retorna el mismo estado recibido, sin modificaciones.

    Nota:
        Esta función puede ser extendida más adelante para realizar validaciones o guardar información en el store.
    """
    # Extrae el mensaje del usuario desde el estado
    user_message = state['user_message']  # Esto se podría reemplazar por un input manual si fuera necesario

    # Retorna el mismo estado original, no cambia nada
    return {
        **state,
    }


def create_itinerary(state: PlannerState, config: RunnableConfig, *, store: BaseStore) -> PlannerState:
    """
    Genera un itinerario de viaje usando los intereses del usuario y su historial de conversación.

    Args:
        state (PlannerState): Estado actual que incluye 'city' y 'user_message'.
        config (RunnableConfig): Configuración de ejecución, contiene identificadores únicos como 'thread_id'.
        store (BaseStore): Almacén de datos donde se guarda y consulta el historial de la conversación.

    Returns:
        PlannerState: Estado actualizado con el itinerario generado en el campo 'itinerary'.

    Flujo de trabajo:
        1. Recupera historial de chat previo desde el almacenamiento (store).
        2. Llama al modelo LLM para generar un itinerario basado en el historial + nuevo mensaje.
        3. Guarda la nueva interacción (mensaje humano + respuesta AI) en el almacenamiento.
    """
    # Crea un identificador único de usuario basado en el thread_id
    user_u = f"user_id_{config['configurable']['thread_id']}"

    # Define un espacio de nombres para almacenar los mensajes
    namespace_u = ("chat_messages", user_u)

    # Recupera el historial de conversación previo (si existe) del store
    store_item = store.get(namespace=namespace_u, key=user_u)
    
    # Extrae los mensajes previos si existen, o inicia una lista vacía
    chat_history_messages = store_item.value['data'] if store_item else []

    # (Debug) Imprime el ID de usuario y su historial de mensajes anterior
    print(user_u, chat_history_messages)

    # Invoca al LLM para generar el itinerario usando ciudad, mensaje actual y historial
    response = llm.invoke(
        itinerary_prompt.format_messages(
            city=state['city'],
            user_message=state['user_message'],
            chat_history=chat_history_messages
        )
    )

    # (Debug) Imprime el contenido del itinerario generado
    print("\nFinal Itinerary:")
    print(response.content)

    # Actualiza el historial en el almacén con el nuevo mensaje humano y respuesta de la AI
    store.put(
        namespace=namespace_u,
        key=user_u,
        value={
            "data": chat_history_messages + [
                HumanMessage(content=state['user_message']),
                AIMessage(content=response.content)
            ]
        }
    )
    
    # Devuelve el estado original, agregando el itinerario generado
    return {
        **state,
        "itinerary": response.content
    }
````

## Definición y Compilación del Workflow con Persistencia en Memoria

````python
# ==========================================================
# CREACIÓN DEL WORKFLOW Y CONFIGURACIÓN DEL ALMACÉN EN MEMORIA
# ==========================================================

# Crear una instancia de almacenamiento en memoria personalizada
# Este almacén guardará el historial de mensajes o el estado entre nodos
in_memory_store_n = CustomMemoryStore(InMemoryStore())

# Crear una instancia de un StateGraph, que es el grafo que maneja el flujo de estados
workflow = StateGraph(PlannerState)

# Agregar nodos al flujo de trabajo:
# Cada nodo representa una función o tarea específica que se ejecutará.
workflow.add_node("input_interests", input_interests)
workflow.add_node("create_itinerary", create_itinerary)

# Definir cuál es el primer nodo que se ejecuta cuando empieza el flujo
workflow.set_entry_point("input_interests")

# Definir las transiciones entre nodos:
# Después de 'input_interests', se ejecuta 'create_itinerary'
workflow.add_edge("input_interests", "create_itinerary")

# Definir que después de 'create_itinerary' el flujo termina (END)
workflow.add_edge("create_itinerary", END)

# Compilar el flujo de trabajo y asociarlo al almacenamiento en memoria
# Esto permite que el flujo recuerde y gestione el estado entre ejecuciones
app = workflow.compile(store=in_memory_store_n)
````

```python
from pymongo import MongoClient
from bson.objectid import ObjectId

class MongoDBStore(BaseStore):
    def __init__(self, db_url: str, db_name: str):
        # Conectarse a MongoDB
        self.client = MongoClient(db_url)
        self.db = self.client[db_name]
        self.collection = self.db['chat_messages']

    def put(self, namespace: tuple, key: str, value: dict):
        # Convertimos el namespace a un nombre de colección y documento
        collection_name = namespace[0]
        document = {"_id": key, "data": value['data']}
        # Guardamos el documento
        self.collection.insert_one(document)

    def get(self, namespace: tuple, key: str):
        # Obtenemos el documento basado en la clave (id)
        document = self.collection.find_one({"_id": key})
        return document

# Ejemplo de uso:
mongo_store = MongoDBStore("mongodb://localhost:27017/", "travel_planner_db")

mongo_store.put(namespace_u, key_u, {"data": ["list a"]})

# Recuperamos el mensaje
item_u = mongo_store.get(namespace_u, key_u)
print(item_u)  # Imprime el documento con los mensajes almacenados
```

#### Explicación:

1. **`MongoDBStore`**: Esta clase se conecta a MongoDB, y ofrece los métodos `put` y `get` para almacenar y recuperar mensajes, respectivamente.
2. **`put`**: Guarda un documento en la colección de MongoDB. El documento incluye la clave y los mensajes.
3. **`get`**: Recupera el documento de MongoDB usando la clave.

#### Paso 1: Instalar dependencias

#### Paso 2: Reemplazar el `CustomMemoryStore` por Redis

Aquí tienes un ejemplo de cómo puedes usar Redis para almacenar y recuperar los mensajes en lugar de usar el `CustomMemoryStore`.

```python
import redis
import json

class RedisStore(BaseStore):
    def __init__(self, host: str = 'localhost', port: int = 6379):
        # Conectarse a Redis
        self.client = redis.StrictRedis(host=host, port=port, db=0, decode_responses=True)

    def put(self, namespace: tuple, key: str, value: dict):
        # Redis utiliza claves simples, por lo que creamos una clave basada en el namespace y la key
        redis_key = f"{namespace[0]}:{key}"
        # Almacenamos el valor en Redis. Los valores deben ser convertidos a JSON.
        self.client.set(redis_key, json.dumps(value))

    def get(self, namespace: tuple, key: str):
        # Generamos la clave Redis
        redis_key = f"{namespace[0]}:{key}"
        # Recuperamos el valor de Redis y lo convertimos de JSON a diccionario
        value = self.client.get(redis_key)
        return json.loads(value) if value else None

# Ejemplo de uso:
redis_store = RedisStore()

# Almacenamos un mensaje
namespace_u = ("chat_messages", "user_id_1")
key_u = "user_id_1"
redis_store.put(namespace_u, key_u, {"data": ["list a"]})

# Recuperamos el mensaje
item_u = redis_store.get(namespace_u, key_u)
print(item_u)  # Imprime el valor almacenado en Redis
```

#### Explicación:

1. **`RedisStore`**: Esta clase se conecta a Redis y ofrece métodos para almacenar (`put`) y recuperar (`get`) datos.
2. **`put`**: Guarda los datos en Redis usando una clave única construida a partir del namespace y la key. El valor se convierte a formato JSON para almacenar objetos complejos.
3. **`get`**: Recupera el valor desde Redis, convirtiéndolo de JSON a un diccionario en Python.

---

### **¿Cuál base de datos elegir?**
- **MongoDB** es una base de datos orientada a documentos, muy útil para almacenar objetos complejos y grandes volúmenes de datos. Es una opción más adecuada si necesitas consultas más avanzadas y estructuras de datos más complejas.
  
- **Redis** es una base de datos en memoria clave-valor, mucho más rápida que MongoDB, pero no está pensada para grandes volúmenes de datos o estructuras de datos complejas. Redis es ideal para almacenamiento temporal o caché.

### **Sustitución en el Workflow**

Una vez que hayas adaptado tu base de datos, el código que antes utilizaba `CustomMemoryStore` para almacenar y recuperar información de la memoria, ahora usará el **`MongoDBStore`** o **`RedisStore`** en lugar de `CustomMemoryStore`. En el flujo de trabajo (`workflow`), la interfaz no cambia; solo cambiarías el almacenamiento de memoria.

Ejemplo:

```python
# Usando MongoDB o Redis en lugar de CustomMemoryStore

# Pasar el store a los nodos en el workflow
app = workflow.compile(store=store)
```

De este modo, puedes sustituir el gestor de memoria local por una base de datos como MongoDB o Redis, y seguir usando el mismo flujo de trabajo.

# 📚 Contexto General

En sistemas basados en agentes (`agents`) como los de LangChain o LangGraph, **la memoria** es usada para guardar **el historial de conversación** (mensajes de usuario y respuestas del agente) para que el agente pueda recordar lo que pasó anteriormente.

**Antes**, estábamos viendo cómo guardar la memoria manualmente usando un **`BaseStore`** personalizado (ya sea con `CustomMemoryStore`, MongoDB o Redis).

**Ahora**, lo que introduces es un mecanismo más "automático" y elegante usando `RunnableWithMessageHistory`, que **automáticamente** maneja la memoria **por cada agente** o **por cada sesión de usuario**, SIN que tengas que escribir tanto código manualmente en cada paso.

---

# 🔥 ¿Qué es `RunnableWithMessageHistory`?

- Es una envoltura (**wrapper**) que puedes poner alrededor de un **runnable** (modelo, agente, flujo de trabajo).
- Su objetivo es **agregar** o **inyectar** el historial de mensajes en cada ejecución automáticamente.
- Además, gestiona dónde y cómo se almacena el historial: puede ser en memoria (`InMemoryChatMessageHistory`) o en una base de datos como Redis, Mongo, etc.

En pocas palabras:

| Antes | Con `RunnableWithMessageHistory` |
|:------|:---------------------------------|
| Guardabas y cargabas el historial manualmente usando `.put` y `.get` en un store | El historial se maneja automáticamente. El runnable ya sabe cómo recordar y actualizar la conversación. |



````python
# Importar las bibliotecas necesarias de LangChain
from langchain_core.chat_history import InMemoryChatMessageHistory
from langchain_core.output_parsers import StrOutputParser
from langchain_core.prompts import ChatPromptTemplate
from langchain_core.runnables.history import RunnableWithMessageHistory
import boto3



# Plantilla de prompt para crear el itinerario de un viaje
itinerary_prompt = ChatPromptTemplate.from_messages([
    ("system", """Eres un asistente de viajes útil. Crea un itinerario de un día para {city} basado en los intereses del usuario. 
    Sigue estas instrucciones:
    1. Usa la conversación a continuación y la última entrada del humano para obtener los intereses del usuario.
    2. Siempre ten en cuenta el tiempo de viaje y los tiempos de comida. Si no es posible hacer todo, dilo.
    3. Si el usuario no ha indicado la época del año o temporada, asume que es verano en {city} y menciona esta suposición en tu respuesta.
    4. Si el usuario no ha indicado un presupuesto de viaje, asume una cantidad razonable en dólares y menciona esta suposición en tu respuesta.
    5. Proporciona un itinerario breve y en viñetas en orden cronológico con horas específicas del día."""),
    MessagesPlaceholder("chat_history"),  # Mantiene el historial de mensajes en el flujo
    ("human", "{user_message}"),  # El mensaje del usuario para personalizar la respuesta
])

````

````python
# Encadenar el prompt con el modelo de lenguaje
chain = itinerary_prompt | llm

# Inicializar el historial de mensajes en memoria
history = InMemoryChatMessageHistory()

# Función para obtener el historial de mensajes
def get_history():
    return history

# Encapsular la cadena en un Runnable que mantenga el historial de mensajes
wrapped_chain = RunnableWithMessageHistory(
    chain,
    get_history,
    history_messages_key="chat_history",  # Clave para acceder al historial de mensajes
)
````

````python
class PlannerState(TypedDict):
    itinerary: str
    city: str
    user_message: str

def input_interests(state: PlannerState, config: RunnableConfig, *, store: BaseStore) -> PlannerState:
    user_message = state['user_message'] #input("Your input: ")
    return {
        **state,
    }

def create_itinerary(state: PlannerState, config: RunnableConfig, *, store: BaseStore) -> PlannerState:
    #- each agent manages it's memory
    response = wrapped_chain.invoke({"city": state['city'], "user_message": state['user_message'], "input": state['user_message']} )
    print("\nFinal Itinerary:")
    print(response.content)
    
    return {
        **state,
        "itinerary": response.content
    }
````
