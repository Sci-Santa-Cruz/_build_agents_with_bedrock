### Introducci√≥n a los Agentes
Los Agentes son sistemas inteligentes o componentes que utilizan Modelos de Lenguaje Grandes (LLMs) para realizar tareas de manera din√°mica y aut√≥noma. Aqu√≠ est√° un desglose de los conceptos clave:

#### ¬øQu√© son los Agentes?
- **Pensamiento paso a paso**: Los agentes aprovechan los LLMs para pensar y razonar sobre problemas de manera estructurada, a menudo denominada razonamiento en cadena de pensamientos. Esto les permite planificar, evaluar y ejecutar tareas de manera efectiva.
- **Acceso a Herramientas**: Los agentes pueden utilizar herramientas externas (por ejemplo, calculadoras, bases de datos, APIs) para mejorar su capacidad de toma de decisiones y resoluci√≥n de problemas.
- **Acceso a Memoria**: Los agentes pueden almacenar y recuperar contexto, lo que les permite trabajar en tareas a lo largo del tiempo, adaptarse a las interacciones con el usuario y manejar flujos de trabajo complejos.

### Caracter√≠sticas clave de los Agentes de IA
- **Percepci√≥n**: La capacidad de recopilar informaci√≥n del entorno a trav√©s de sensores o entradas de datos.
- **Toma de decisiones**: El uso de algoritmos de IA para procesar la informaci√≥n y determinar la mejor acci√≥n a tomar.
- **Acci√≥n**: La capacidad de ejecutar decisiones e interactuar con el entorno o los usuarios.
- **Aprendizaje**: La capacidad de mejorar el rendimiento a lo largo del tiempo mediante la experiencia y la retroalimentaci√≥n.
- **Autonom√≠a**: Operar de manera independiente hasta cierto punto, sin intervenci√≥n humana constante.
- **Orientaci√≥n a objetivos**: Trabajar hacia objetivos o tareas espec√≠ficas.

## Detalles del Caso de Uso

Nuestro Planificador de Viajes sigue un proceso sencillo de tres pasos:

### Entrada Inicial del Usuario:
- La aplicaci√≥n solicita al usuario que ingrese su plan de viaje deseado para obtener asistencia del Agente de IA.
- Esta informaci√≥n se almacena en el estado.

### Entrada de Intereses:

- Estos intereses se almacenan como una lista en el estado.

### Creaci√≥n del Itinerario:

El flujo entre estos pasos es gestionado por LangGraph, que maneja las transiciones de estado y asegura que cada paso se ejecute en el orden correcto.


## Conceptos B√°sicos de LangGraph

### Componentes Clave

#### StateGraph
- Este objeto encapsula el grafo que se est√° recorriendo durante la ejecuci√≥n.
- Es el n√∫cleo de nuestra aplicaci√≥n, definiendo el flujo de nuestro Planificador de Viajes.
- **PlannerState**: Un tipo personalizado que representa el estado de nuestro proceso de planificaci√≥n.

#### Nodos

En LangGraph, los nodos son t√≠picamente funciones de Python. 
Hay dos nodos principales que usaremos en nuestro grafo:

1. **Nodo del agente**: Es responsable de decidir qu√© acciones tomar (si es que se deben tomar).
2. **Nodo de herramienta**: Este nodo orquesta la llamada a la herramienta respectiva y devuelve la salida.

#### Aristas

Las aristas definen c√≥mo se enruta la l√≥gica y c√≥mo el grafo decide cu√°ndo detenerse.
Estas definen c√≥mo trabajan los agentes y c√≥mo los diferentes nodos se comunican entre s√≠.
Algunos tipos clave de aristas son:

- **Aristas normales**: Van directamente de un nodo a otro.
- **Aristas condicionales**: Llaman a una funci√≥n para determinar a qu√© nodo(s) ir a continuaci√≥n.
- **Punto de entrada**: Define qu√© nodo se debe llamar primero cuando llega la entrada del usuario.
- **Punto de entrada condicional**: Llama a una funci√≥n para determinar qu√© nodo(s) se debe llamar primero cuando llega la entrada del usuario.

````python
import os
from typing import TypedDict, Annotated, List
from langgraph.graph import StateGraph, END
from langchain_core.messages import HumanMessage, AIMessage
from langchain_core.prompts import ChatPromptTemplate
from langchain_core.runnables.graph import MermaidDrawMethod
from IPython.display import display, Image
from dotenv import load_dotenv
import os
#load_dotenv()
````

````python
class PlannerState(TypedDict):
    messages: Annotated[List[HumanMessage | AIMessage], "The messages in the conversation"]
    itinerary: str
    city: str
    user_message: str
````


## Configuraci√≥n del Modelo de Lenguaje y los Prompts



````python
from langchain_aws import ChatBedrockConverse
import boto3
from langgraph.checkpoint.memory import MemorySaver
from langchain_core.runnables.config import RunnableConfig
from langchain_core.chat_history import InMemoryChatMessageHistory
from langchain_core.output_parsers import StrOutputParser
from langchain_core.prompts import ChatPromptTemplate
from langchain_core.runnables.history import RunnableWithMessageHistory
from langchain_core.prompts import ChatPromptTemplate, MessagesPlaceholder

# ---- ‚ö†Ô∏è Actualiza la regi√≥n para tu configuraci√≥n de AWS ‚ö†Ô∏è ----
region = 'us-east-2'

bedrock = boto3.client(
    service_name='bedrock-runtime',
    region_name=region,
)

bedrock_client = boto3.client("bedrock-runtime", region_name=region)

model_id = "us.anthropic.claude-3-5-haiku-20241022-v1:0"

llm = ChatBedrockConverse(
    model=model_id,
    temperature=0,
    max_tokens=5000,
    client=bedrock_client,
)

itinerary_prompt = ChatPromptTemplate.from_messages([
    ("system", """Eres un asistente de viaje √∫til. Crea un itinerario para un viaje de un d√≠a en {city} basado en los intereses del usuario.
    Sigue estas instrucciones:
    1. Usa la conversaci√≥n de chat a continuaci√≥n y la √∫ltima entrada del humano para obtener los intereses del usuario.
    2. Siempre ten en cuenta el tiempo de viaje y los tiempos de comida: si no es posible hacer todo, dilo.
    3. Si el usuario no ha mencionado una √©poca del a√±o o temporada, asume que es verano en {city} y menciona esta suposici√≥n en tu respuesta.
    4. Si el usuario no ha mencionado un presupuesto de viaje, asume una cantidad razonable en d√≥lares y menciona esta suposici√≥n en tu respuesta.
    5. Proporciona un itinerario breve y en vi√±etas en orden cronol√≥gico con las horas espec√≠ficas del d√≠a."""), 
    MessagesPlaceholder("chat_history"),
    ("human", "{user_message}"),
])
````



## **Definir los nodos y las aristas**  


Estamos a√±adiendo los nodos, las aristas, as√≠ como una memoria persistente al **StateGraph** antes de compilarlo.

1. **Planes de viaje del usuario**  
2. **Invocar con Bedrock**  
3. **Generar el plan de viaje para el d√≠a**  
4. **Capacidad de agregar o modificar el plan**



````python
def input_interests(state: PlannerState) -> PlannerState:
    user_message = state['user_message']  # input("Tu mensaje: ")
    # print(f"Vamos a :: {user_message}:: para el viaje a {state['city']} basado en tus intereses mencionados en el mensaje....")

    if not state.get('messages', None):
        state['messages'] = []
    return {
        **state,
    }

def create_itinerary(state: PlannerState) -> PlannerState:
    response = llm.invoke(itinerary_prompt.format_messages(city=state['city'],
                         user_message=state['user_message'],
                         chat_history=state['messages']))
    print("\nItinerario Final:")
    print(response.content)
    return {
        **state,
        "messages": state['messages'] + [HumanMessage(content=state['user_message']), AIMessage(content=response.content)],
        "itinerary": response.content
    }
````


## **Crear y compilar el gr√°fico**

Ahora crearemos nuestro flujo de trabajo en LangGraph y lo compilaremos.

Primero, inicializamos un **StateGraph** con la clase **State** que definimos anteriormente.
Luego, agregamos nuestros **nodos** y **aristas** (edges).
Usamos el **Nodo INICIO** (*START Node*), un nodo especial que env√≠a la entrada del usuario al gr√°fico, para indicar d√≥nde comenzar nuestro gr√°fico.
El **Nodo FINAL** (*END Node*) es un nodo especial que representa un nodo terminal.


````python
# Inicializa el grafo de estados con un tipo de estado espec√≠fico, en este caso, PlannerState
workflow = StateGraph(PlannerState)

# Agrega nodos al grafo de flujo de trabajo que representan diferentes etapas del proceso
# Cada nodo ser√° una funci√≥n o tarea que el grafo de estados ejecutar√°.
workflow.add_node("input_interests", input_interests)  # Nodo que maneja los intereses del usuario
workflow.add_node("create_itinerary", create_itinerary)  # Nodo que crea el itinerario con la informaci√≥n

# Establece el punto de entrada del grafo como el nodo 'input_interests'
# Esto significa que el flujo de trabajo comenzar√° en este nodo.
workflow.set_entry_point("input_interests")

# Agrega conexiones (aristas) entre nodos, lo que define el flujo de ejecuci√≥n
# En este caso, el flujo de trabajo va de 'input_interests' a 'create_itinerary' y luego termina en el nodo END.
workflow.add_edge("input_interests", "create_itinerary")  # Despu√©s de manejar los intereses, pasa a crear el itinerario
workflow.add_edge("create_itinerary", END)  # Finalmente, el flujo termina

# El checkpointer permite que el grafo guarde su estado, persistiendo la memoria del flujo de trabajo.
# Esto es √∫til para poder recuperar y continuar desde el mismo punto en caso de reinicio o fallo.
memory = MemorySaver()  # Crea una instancia del MemorySaver para almacenar el estado del flujo de trabajo

# Compila el grafo de trabajo en una aplicaci√≥n, pasando el checkpointer (MemorySaver) para que se guarde el estado del flujo de trabajo
app = workflow.compile(checkpointer=memory)
````


## **Mostrar la estructura del gr√°fico**

````python
def run_travel_planner(user_request: str, config_dict: dict):
    print(f"Solicitud actual del usuario: {user_request}\n")
    init_input = {"user_message": user_request, "city": "Seattle"}

    for output in app.stream(init_input, config=config_dict, stream_mode="values"):
        pass  # Los nodos ahora manejan toda la impresi√≥n por s√≠ mismos
````

### Crear una persistencia externa de memoria

En esta secci√≥n aprovecharemos la persistencia de m√∫ltiples hilos y sesiones para los mensajes de Chat. Lo ideal es usar soluciones de persistencia como Redis Store, etc., para guardar mensajes por sesi√≥n.

### Gesti√≥n de Memoria

Existen varios patrones para la gesti√≥n de memoria:

1. Cada agente puede tener su propia memoria de sesi√≥n.
2. La memoria de todo el gr√°fico puede ser combinada, donde cada agente tendr√° su propia memoria.

El `MemorySaver` o el Store tienen el concepto de separar secciones de memoria mediante *Namespaces* o *Thread ID's*, y estos pueden ser aprovechados para:

1. Usar la memoria a nivel de gr√°fico.
2. Cada agente puede tener su propia memoria a trav√©s de espacio en el "saver" o tener su propio "saver", como lo hacemos en el agente ReACT.


<img src="images/multi_memory_light.png" width="800" height="600">


````python
# Importaci√≥n de las clases necesarias para el funcionamiento de la memoria personalizada
from langgraph.store.base import BaseStore, Item, Op, Result
from langgraph.store.memory import InMemoryStore
from typing import Any, Iterable, Literal, NamedTuple, Optional, Union, cast

# Definici√≥n de la clase CustomMemoryStore que hereda de BaseStore
class CustomMemoryStore(BaseStore):
    """
    Esta clase extiende la funcionalidad de almacenamiento de la clase BaseStore
    permitiendo el uso de un almac√©n externo para gestionar datos.
    
    Acepta un 'ext_store' (almac√©n externo) que puede ser cualquier clase que
    implemente la interfaz del BaseStore.
    """

    def __init__(self, ext_store: BaseStore):
        """
        Inicializa el objeto CustomMemoryStore con un almac√©n externo.

        Args:
            ext_store (BaseStore): Almac√©n externo que se usar√° para la gesti√≥n de datos.
        """
        # El almac√©n proporcionado se guarda como un atributo de la instancia
        self.store = ext_store

    def get(self, namespace: tuple[str, ...], key: str) -> Optional[Item]:
        """
        Recupera un elemento del almac√©n usando un espacio de nombres y una clave.

        Args:
            namespace (tuple[str, ...]): Tupla que define el espacio de nombres para el almacenamiento.
            key (str): Clave del item que se desea obtener.

        Returns:
            Optional[Item]: El √≠tem recuperado del almac√©n, o None si no existe.
        """
        # Llama al m√©todo 'get' del almac√©n externo para obtener el √≠tem
        return self.store.get(namespace, key)

    def put(self, namespace: tuple[str, ...], key: str, value: dict[str, Any]) -> None:
        """
        Almacena un √≠tem en el almac√©n con un espacio de nombres y clave dados.

        Args:
            namespace (tuple[str, ...]): Tupla que define el espacio de nombres para el almacenamiento.
            key (str): Clave bajo la cual se almacenar√° el √≠tem.
            value (dict[str, Any]): El valor del √≠tem a almacenar.

        Returns:
            None: El m√©todo no devuelve nada.
        """
        # Llama al m√©todo 'put' del almac√©n externo para almacenar el valor
        return self.store.put(namespace, key, value)

    async def aget(self, namespace: tuple[str, ...], key: str) -> Optional[Item]:
        """
        Recupera de forma as√≠ncrona un elemento del almac√©n usando un espacio de nombres y una clave.

        Args:
            namespace (tuple[str, ...]): Tupla que define el espacio de nombres para el almacenamiento.
            key (str): Clave del item que se desea obtener.

        Returns:
            Optional[Item]: El √≠tem recuperado del almac√©n, o None si no existe.
        """
        # Llama al m√©todo 'get' de forma as√≠ncrona
        res = await self.get(namespace, key)
        return res

    async def aput(self, namespace: tuple[str, ...], key: str, value: dict[str, Any]) -> None:
        """
        Almacena de forma as√≠ncrona un √≠tem en el almac√©n con un espacio de nombres y clave dados.

        Args:
            namespace (tuple[str, ...]): Tupla que define el espacio de nombres para el almacenamiento.
            key (str): Clave bajo la cual se almacenar√° el √≠tem.
            value (dict[str, Any]): El valor del √≠tem a almacenar.

        Returns:
            None: El m√©todo no devuelve nada.
        """
        # Llama al m√©todo 'put' de forma as√≠ncrona
        res = await self.put(namespace, key, value)
        return res

    def batch(self, ops: Iterable[Op]) -> list[Result]:
        """
        Realiza m√∫ltiples operaciones de almacenamiento en batch (lote).

        Args:
            ops (Iterable[Op]): Conjunto de operaciones de almacenamiento que se desean realizar.

        Returns:
            list[Result]: Lista de resultados de cada operaci√≥n realizada.
        """
        # Llama al m√©todo 'batch' del almac√©n externo para ejecutar las operaciones en batch
        return self.store.batch(ops)

    async def abatch(self, ops: Iterable[Op]) -> list[Result]:
        """
        Realiza m√∫ltiples operaciones de almacenamiento en batch de forma as√≠ncrona.

        Args:
            ops (Iterable[Op]): Conjunto de operaciones de almacenamiento que se desean realizar.

        Returns:
            list[Result]: Lista de resultados de cada operaci√≥n realizada.
        """
        # Llama al m√©todo 'abatch' del almac√©n externo para ejecutar las operaciones en batch de forma as√≠ncrona
        res = await self.store.abatch(ops)
        return res
````



### Explicaci√≥n de la clase `CustomMemoryStore`:

1. **Clase `CustomMemoryStore`**:
   - Esta clase extiende `BaseStore` y utiliza un almac√©n externo (`ext_store`) que puede ser un almac√©n en memoria o cualquier otro tipo de almac√©n que implemente la interfaz `BaseStore`.
   
2. **M√©todos principales**:
   - `get`: Recupera un √≠tem del almac√©n externo usando un espacio de nombres y una clave.
   - `put`: Almacena un √≠tem en el almac√©n externo usando un espacio de nombres y una clave.
   - `aget` y `aput`: Son las versiones as√≠ncronas de `get` y `put`, utilizadas cuando se manejan operaciones as√≠ncronas.
   - `batch`: Realiza m√∫ltiples operaciones en lote de manera sincr√≥nica.
   - `abatch`: Realiza m√∫ltiples operaciones en lote de manera as√≠ncrona.

### Uso de las operaciones de batch:
- **Batch** permite ejecutar varias operaciones de almacenamiento (como insertar, actualizar o eliminar elementos) de manera eficiente en un solo proceso. Esto es √∫til para optimizar el rendimiento cuando se necesitan realizar m√∫ltiples acciones de forma consecutiva.

### Asincron√≠a:
- Los m√©todos `aget` y `aput`, junto con `abatch`, permiten realizar operaciones de almacenamiento de manera as√≠ncrona, lo que puede ser √∫til en aplicaciones que requieren no bloquear el hilo principal mientras se espera la operaci√≥n de almacenamiento.

Este c√≥digo est√° dise√±ado para ofrecer flexibilidad al trabajar con almacenamiento de memoria de manera tanto sincr√≥nica como asincr√≥nica.

````python
# 1. Creamos un almac√©n de memoria temporal.
# Usamos CustomMemoryStore que utiliza InMemoryStore internamente (es decir, guarda en RAM).
in_memory_store = CustomMemoryStore(InMemoryStore())

# 2. Definimos un espacio de nombres (namespace) para agrupar datos relacionados.
# Aqu√≠ agrupamos los mensajes de chat de un usuario.
namespace_u = ("chat_messages", "user_id_1")

# 3. Definimos una clave (key) √∫nica que identifica a un usuario espec√≠fico
key_u = "user_id_1" 

# 4. Definimos el contenido que queremos guardar.
# Es un diccionario que contiene una lista con elementos 
data_to_store = {
    "data": ["hola que tal", "dame un plan a los maiamis", "agrega visita a Little Havana"]
}

# 5. Guardamos los datos en el almac√©n de memoria.
# Se guarda usando el namespace y la key definidos anteriormente.
in_memory_store.put(namespace_u, key_u, data_to_store)

# 6. Recuperamos el dato que acabamos de guardar, utilizando el mismo namespace y key.
item_u = in_memory_store.get(namespace_u, key_u)


````



### **Crear un gr√°fico similar al anterior** 

‚Äî Ten en cuenta que no tendremos mensajes en el estado del gr√°fico ya que estos han sido externalizados.



````python
# Definir el estado del planificador como un TypedDict
class PlannerState(TypedDict):
    itinerary: str  # Itinerario del viaje
    city: str       # Ciudad para la que se est√° creando el itinerario
    user_message: str  # Mensaje original del usuario
````

````python
def input_interests(state: PlannerState, config: RunnableConfig, *, store: BaseStore) -> PlannerState:
    """
    Procesa el primer mensaje del usuario que contiene sus intereses de viaje.

    Args:
        state (PlannerState): Estado actual del flujo que contiene, entre otros, el 'user_message'.
        config (RunnableConfig): Configuraci√≥n adicional de ejecuci√≥n, contiene identificadores como 'thread_id'.
        store (BaseStore): Almac√©n de datos para persistir o consultar informaci√≥n (no utilizado en esta funci√≥n).

    Returns:
        PlannerState: Retorna el mismo estado recibido, sin modificaciones.

    Nota:
        Esta funci√≥n puede ser extendida m√°s adelante para realizar validaciones o guardar informaci√≥n en el store.
    """
    # Extrae el mensaje del usuario desde el estado
    user_message = state['user_message']  # Esto se podr√≠a reemplazar por un input manual si fuera necesario

    # Retorna el mismo estado original, no cambia nada
    return {
        **state,
    }


def create_itinerary(state: PlannerState, config: RunnableConfig, *, store: BaseStore) -> PlannerState:
    """
    Genera un itinerario de viaje usando los intereses del usuario y su historial de conversaci√≥n.

    Args:
        state (PlannerState): Estado actual que incluye 'city' y 'user_message'.
        config (RunnableConfig): Configuraci√≥n de ejecuci√≥n, contiene identificadores √∫nicos como 'thread_id'.
        store (BaseStore): Almac√©n de datos donde se guarda y consulta el historial de la conversaci√≥n.

    Returns:
        PlannerState: Estado actualizado con el itinerario generado en el campo 'itinerary'.

    Flujo de trabajo:
        1. Recupera historial de chat previo desde el almacenamiento (store).
        2. Llama al modelo LLM para generar un itinerario basado en el historial + nuevo mensaje.
        3. Guarda la nueva interacci√≥n (mensaje humano + respuesta AI) en el almacenamiento.
    """
    # Crea un identificador √∫nico de usuario basado en el thread_id
    user_u = f"user_id_{config['configurable']['thread_id']}"

    # Define un espacio de nombres para almacenar los mensajes
    namespace_u = ("chat_messages", user_u)

    # Recupera el historial de conversaci√≥n previo (si existe) del store
    store_item = store.get(namespace=namespace_u, key=user_u)
    
    # Extrae los mensajes previos si existen, o inicia una lista vac√≠a
    chat_history_messages = store_item.value['data'] if store_item else []

    # (Debug) Imprime el ID de usuario y su historial de mensajes anterior
    print(user_u, chat_history_messages)

    # Invoca al LLM para generar el itinerario usando ciudad, mensaje actual y historial
    response = llm.invoke(
        itinerary_prompt.format_messages(
            city=state['city'],
            user_message=state['user_message'],
            chat_history=chat_history_messages
        )
    )

    # (Debug) Imprime el contenido del itinerario generado
    print("\nFinal Itinerary:")
    print(response.content)

    # Actualiza el historial en el almac√©n con el nuevo mensaje humano y respuesta de la AI
    store.put(
        namespace=namespace_u,
        key=user_u,
        value={
            "data": chat_history_messages + [
                HumanMessage(content=state['user_message']),
                AIMessage(content=response.content)
            ]
        }
    )
    
    # Devuelve el estado original, agregando el itinerario generado
    return {
        **state,
        "itinerary": response.content
    }
````

## Definici√≥n y Compilaci√≥n del Workflow con Persistencia en Memoria

````python
# ==========================================================
# CREACI√ìN DEL WORKFLOW Y CONFIGURACI√ìN DEL ALMAC√âN EN MEMORIA
# ==========================================================

# Crear una instancia de almacenamiento en memoria personalizada
# Este almac√©n guardar√° el historial de mensajes o el estado entre nodos
in_memory_store_n = CustomMemoryStore(InMemoryStore())

# Crear una instancia de un StateGraph, que es el grafo que maneja el flujo de estados
workflow = StateGraph(PlannerState)

# Agregar nodos al flujo de trabajo:
# Cada nodo representa una funci√≥n o tarea espec√≠fica que se ejecutar√°.
workflow.add_node("input_interests", input_interests)
workflow.add_node("create_itinerary", create_itinerary)

# Definir cu√°l es el primer nodo que se ejecuta cuando empieza el flujo
workflow.set_entry_point("input_interests")

# Definir las transiciones entre nodos:
# Despu√©s de 'input_interests', se ejecuta 'create_itinerary'
workflow.add_edge("input_interests", "create_itinerary")

# Definir que despu√©s de 'create_itinerary' el flujo termina (END)
workflow.add_edge("create_itinerary", END)

# Compilar el flujo de trabajo y asociarlo al almacenamiento en memoria
# Esto permite que el flujo recuerde y gestione el estado entre ejecuciones
app = workflow.compile(store=in_memory_store_n)
````

````python
# ==========================================================
# FUNCI√ìN PARA EJECUTAR EL FLUJO DE TRAVEL PLANNER
# ==========================================================

def run_travel_planner(user_request: str, config_dict: dict):
    """
    Ejecuta el flujo del planificador de viajes (Travel Planner).

    Par√°metros:
    - user_request (str): El mensaje o solicitud que el usuario ingresa (por ejemplo, "Me gustan los museos y la comida local").
    - config_dict (dict): Diccionario de configuraci√≥n para el flujo (ej. contiene IDs de usuario, par√°metros de memoria, etc.).

    Proceso:
    - Inicializa el input del flujo con el mensaje del usuario y una ciudad fija (en este caso, "Seattle").
    - Ejecuta el flujo de trabajo (workflow `app`) transmitiendo los estados a trav√©s de los nodos definidos.
    - Los nodos son responsables de procesar y mostrar sus salidas. Esta funci√≥n solo inicia el flujo.
    """
    
    # Mostrar por consola la solicitud actual del usuario
    print(f"Current User Request: {user_request}\n")

    # Crear el input inicial para el flujo, con el mensaje del usuario y la ciudad predefinida
    init_input = {
        "user_message": user_request,
        # Puedes cambiar esto din√°micamente si deseas
        "city": "Seattle"  # NO se usa
    }

    # Ejecutar el flujo de trabajo (workflow)
    # 'app.stream()' procesa el flujo nodo por nodo en modo "stream" (flujo continuo de datos)
    for output in app.stream(init_input, config=config_dict, stream_mode="values"):
        pass  # No hacemos nada aqu√≠ porque los nodos ya imprimen los resultados dentro de s√≠ mismos


````

```python
from pymongo import MongoClient
from bson.objectid import ObjectId

class MongoDBStore(BaseStore):
    def __init__(self, db_url: str, db_name: str):
        # Conectarse a MongoDB
        self.client = MongoClient(db_url)
        self.db = self.client[db_name]
        self.collection = self.db['chat_messages']

    def put(self, namespace: tuple, key: str, value: dict):
        # Convertimos el namespace a un nombre de colecci√≥n y documento
        collection_name = namespace[0]
        document = {"_id": key, "data": value['data']}
        # Guardamos el documento
        self.collection.insert_one(document)

    def get(self, namespace: tuple, key: str):
        # Obtenemos el documento basado en la clave (id)
        document = self.collection.find_one({"_id": key})
        return document

# Ejemplo de uso:
mongo_store = MongoDBStore("mongodb://localhost:27017/", "travel_planner_db")

# Almacenamos un mensaje
namespace_u = ("chat_messages", "user_id_1")
key_u = "user_id_1"
mongo_store.put(namespace_u, key_u, {"data": ["list a"]})

# Recuperamos el mensaje
item_u = mongo_store.get(namespace_u, key_u)
print(item_u)  # Imprime el documento con los mensajes almacenados
```

#### Explicaci√≥n:

1. **`MongoDBStore`**: Esta clase se conecta a MongoDB, y ofrece los m√©todos `put` y `get` para almacenar y recuperar mensajes, respectivamente.
2. **`put`**: Guarda un documento en la colecci√≥n de MongoDB. El documento incluye la clave y los mensajes.
3. **`get`**: Recupera el documento de MongoDB usando la clave.

### **2. Usando Redis como base de datos para almacenar los datos**

#### Paso 1: Instalar dependencias

Para interactuar con Redis, necesitas instalar la librer√≠a `redis-py`.


#### Paso 2: Reemplazar el `CustomMemoryStore` por Redis

Aqu√≠ tienes un ejemplo de c√≥mo puedes usar Redis para almacenar y recuperar los mensajes en lugar de usar el `CustomMemoryStore`.

```python
import redis
import json

class RedisStore(BaseStore):
    def __init__(self, host: str = 'localhost', port: int = 6379):
        # Conectarse a Redis
        self.client = redis.StrictRedis(host=host, port=port, db=0, decode_responses=True)

    def put(self, namespace: tuple, key: str, value: dict):
        # Redis utiliza claves simples, por lo que creamos una clave basada en el namespace y la key
        redis_key = f"{namespace[0]}:{key}"
        # Almacenamos el valor en Redis. Los valores deben ser convertidos a JSON.
        self.client.set(redis_key, json.dumps(value))

    def get(self, namespace: tuple, key: str):
        # Generamos la clave Redis
        redis_key = f"{namespace[0]}:{key}"
        # Recuperamos el valor de Redis y lo convertimos de JSON a diccionario
        value = self.client.get(redis_key)
        return json.loads(value) if value else None

# Ejemplo de uso:
redis_store = RedisStore()

# Almacenamos un mensaje
namespace_u = ("chat_messages", "user_id_1")
key_u = "user_id_1"
redis_store.put(namespace_u, key_u, {"data": ["list a"]})

# Recuperamos el mensaje
item_u = redis_store.get(namespace_u, key_u)
print(item_u)  # Imprime el valor almacenado en Redis
```

#### Explicaci√≥n:

1. **`RedisStore`**: Esta clase se conecta a Redis y ofrece m√©todos para almacenar (`put`) y recuperar (`get`) datos.
2. **`put`**: Guarda los datos en Redis usando una clave √∫nica construida a partir del namespace y la key. El valor se convierte a formato JSON para almacenar objetos complejos.
3. **`get`**: Recupera el valor desde Redis, convirti√©ndolo de JSON a un diccionario en Python.

---

### **¬øCu√°l base de datos elegir?**
- **MongoDB** es una base de datos orientada a documentos, muy √∫til para almacenar objetos complejos y grandes vol√∫menes de datos. Es una opci√≥n m√°s adecuada si necesitas consultas m√°s avanzadas y estructuras de datos m√°s complejas.
  
- **Redis** es una base de datos en memoria clave-valor, mucho m√°s r√°pida que MongoDB, pero no est√° pensada para grandes vol√∫menes de datos o estructuras de datos complejas. Redis es ideal para almacenamiento temporal o cach√©.

### **Sustituci√≥n en el Workflow**

Una vez que hayas adaptado tu base de datos, el c√≥digo que antes utilizaba `CustomMemoryStore` para almacenar y recuperar informaci√≥n de la memoria, ahora usar√° el **`MongoDBStore`** o **`RedisStore`** en lugar de `CustomMemoryStore`. En el flujo de trabajo (`workflow`), la interfaz no cambia; solo cambiar√≠as el almacenamiento de memoria.

Ejemplo:

```python
# Usando MongoDB o Redis en lugar de CustomMemoryStore

# Pasar el store a los nodos en el workflow

```

# üìö Contexto General

En sistemas basados en agentes (`agents`) como los de LangChain o LangGraph, **la memoria** es usada para guardar **el historial de conversaci√≥n** (mensajes de usuario y respuestas del agente) para que el agente pueda recordar lo que pas√≥ anteriormente.

**Antes**, est√°bamos viendo c√≥mo guardar la memoria manualmente usando un **`BaseStore`** personalizado (ya sea con `CustomMemoryStore`, MongoDB o Redis).

**Ahora**, lo que introduces es un mecanismo m√°s "autom√°tico" y elegante usando `RunnableWithMessageHistory`, que **autom√°ticamente** maneja la memoria **por cada agente** o **por cada sesi√≥n de usuario**, SIN que tengas que escribir tanto c√≥digo manualmente en cada paso.

---

# üî• ¬øQu√© es `RunnableWithMessageHistory`?

- Es una envoltura (**wrapper**) que puedes poner alrededor de un **runnable** (modelo, agente, flujo de trabajo).
- Su objetivo es **agregar** o **inyectar** el historial de mensajes en cada ejecuci√≥n autom√°ticamente.
- Adem√°s, gestiona d√≥nde y c√≥mo se almacena el historial: puede ser en memoria (`InMemoryChatMessageHistory`) o en una base de datos como Redis, Mongo, etc.

En pocas palabras:

| Antes | Con `RunnableWithMessageHistory` |
|:------|:---------------------------------|
| Guardabas y cargabas el historial manualmente usando `.put` y `.get` en un store | El historial se maneja autom√°ticamente. El runnable ya sabe c√≥mo recordar y actualizar la conversaci√≥n. |



````python
# Importar las bibliotecas necesarias de LangChain
from langchain_core.chat_history import InMemoryChatMessageHistory
from langchain_core.output_parsers import StrOutputParser
from langchain_core.prompts import ChatPromptTemplate
from langchain_core.runnables.history import RunnableWithMessageHistory
import boto3



# Plantilla de prompt para crear el itinerario de un viaje
itinerary_prompt = ChatPromptTemplate.from_messages([
    ("system", """Eres un asistente de viajes √∫til. Crea un itinerario de un d√≠a para {city} basado en los intereses del usuario. 
    Sigue estas instrucciones:
    1. Usa la conversaci√≥n a continuaci√≥n y la √∫ltima entrada del humano para obtener los intereses del usuario.
    2. Siempre ten en cuenta el tiempo de viaje y los tiempos de comida. Si no es posible hacer todo, dilo.
    3. Si el usuario no ha indicado la √©poca del a√±o o temporada, asume que es verano en {city} y menciona esta suposici√≥n en tu respuesta.
    4. Si el usuario no ha indicado un presupuesto de viaje, asume una cantidad razonable en d√≥lares y menciona esta suposici√≥n en tu respuesta.
    5. Proporciona un itinerario breve y en vi√±etas en orden cronol√≥gico con horas espec√≠ficas del d√≠a."""),
    MessagesPlaceholder("chat_history"),  # Mantiene el historial de mensajes en el flujo
    ("human", "{user_message}"),  # El mensaje del usuario para personalizar la respuesta
])

````

````python
# Encadenar el prompt con el modelo de lenguaje
chain = itinerary_prompt | llm

# Inicializar el historial de mensajes en memoria
history = InMemoryChatMessageHistory()

# Funci√≥n para obtener el historial de mensajes
def get_history():
    return history

# Encapsular la cadena en un Runnable que mantenga el historial de mensajes
wrapped_chain = RunnableWithMessageHistory(
    chain,
    get_history,
    history_messages_key="chat_history",  # Clave para acceder al historial de mensajes
)
````

.



````python
class PlannerState(TypedDict):
    itinerary: str
    city: str
    user_message: str

def input_interests(state: PlannerState, config: RunnableConfig, *, store: BaseStore) -> PlannerState:
    user_message = state['user_message'] #input("Your input: ")
    return {
        **state,
    }

def create_itinerary(state: PlannerState, config: RunnableConfig, *, store: BaseStore) -> PlannerState:
    #- each agent manages it's memory
    response = wrapped_chain.invoke({"city": state['city'], "user_message": state['user_message'], "input": state['user_message']} )
    print("\nFinal Itinerary:")
    print(response.content)
    
    return {
        **state,
        "itinerary": response.content
    }
````

````python
workflow = StateGraph(PlannerState)

#workflow.add_node("input_city", input_city)
workflow.add_node("input_interests", input_interests)
workflow.add_node("create_itinerary", create_itinerary)

workflow.set_entry_point("input_interests")

#workflow.add_edge("input_city", "input_interests")
workflow.add_edge("input_interests", "create_itinerary")
workflow.add_edge("create_itinerary", END)


app = workflow.compile()
````

````python
def run_travel_planner(user_request: str, config_dict: dict):
    print(f"Current User Request: {user_request}\n")
    init_input = {"user_message": user_request,"city" : "Seattle"}

    for output in app.stream(init_input, config=config_dict, stream_mode="values"):
        pass  # The nodes themselves now handle all printing

config = {"configurable": {"thread_id": "1"}}

````
